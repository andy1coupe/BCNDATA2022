# cust lifetime value 

```python
full['Customer Lifetime Value'].value_counts()
# deal with nulls 
# if theres a % sign , remove the sign and x 100 
# turn into a float 
# else, leave value as is 
```

```python
full['Customer Lifetime Value']=full['Customer Lifetime Value']*100
```

```python
full['Customer Lifetime Value']=pd.to_numeric(full['Customer Lifetime Value'],errors='coerce')
```

```python
full['Customer Lifetime Value'].describe()
```

```python
full.info()
```

```python
#replace nulls with 0 
full['Customer Lifetime Value'].fillna('0',inplace=True)
```

```python
def clean_clv(x):
    if str(x).endswith('%') is True:
        y=float(x.replace('%',''))*100
        return y
    else:
        return float(x)
```

```python
# apply to full column 
full['Customer Lifetime Value']=full['Customer Lifetime Value'].apply(clean_clv)
```
